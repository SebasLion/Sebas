Costo = 259 * 200 = 51,800

```python
def costo_tina(n):
  l1 = []
  costo=0
  contador = 0
  for k in range(n):
    l2 = []
    for i in range(5):
      u = rnd.triangular (190, 230, 210)
      l2.append(u)
    cantidad = sum(l2)
    if cantidad > 1000:
      contador += 1
      prob = contador/n
      costo=costo+200
      l1.append(prob)
  return l1,costo
```
