  2. En el famoso juego de la ruleta, existen muchas opciones para apostar. Una de ellas consiste en apostarle al color rojo o al color negro. En el tablero de la ruleta existen 10 números rojos, 10 números negros y 2 números verdes (cero y doble cero). Si un jugador apuesta a un color y el color aparece, él o ella gana la cantidad apostada. Si otro color aparece, el jugador pierde la cantidad apostada. Si el color verde aparece, la rueda de la ruleta se vuelve a girar hasta que el color rojo o negro aparezca. Si este color es el color que se apostó, el jugador no gana ni pierde. De otra forma, se pierde la cantidad apostada. Dos jugadores usan diferentes estrategias. Un jugador simplemente apuesta $\$1$ al color rojo cada vez. El otro jugador empieza apostando un $\$1$ al color rojo. Si él gana, él apuesta otro $ \$1$. Sin embargo, si él pierde, él apuesta $\$2$ la próxima vez. Si él pierde otra vez, él apuesta $\$4$. Este jugador puede continuar doblando la apuesta hasta un límite de $\$500$. Si él pierde esta apuesta de $\$500$, empieza apostando nuevamente $\$1$. Si cada jugador inicia el juego con $\$200$, ¿Cuál cree usted que es la mejor estrategia?

Como esto es un ejercicio de simulación, debemos crear números aleatorios $U \sim U(0,1)$

```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rul_roja(n):
  l_rojo=[]

  exito1=0
  for i in range(n):
    R=rnd.uniform(0,1)
    if 0 <= R < 10/22:
      exito1+=1
    prob1 = exito1 / (i + 1)
    l_rojo.append(prob1)
  return l_rojo
```

```python
def rul_negra(n):
  l_negro=[]
  exito2=0
  for i in range(n):
    R=rnd.uniform(0,1)
    if (10 / 22) <= R < (20 / 22):
        exito2 += 1
    prob2 = exito2 / (i + 1)
    l_negro.append(prob2)
  return l_negro
```

```python
def rul_verde(n):
  l_verde=[]
  exito3=0
  for i in range(n):
    R=rnd.uniform(0,1)
    if 20/22 <= R <= 22/22:
      exito3+=1
    prob3=exito3/(i+1)
    l_verde.append(prob3)
  return l_verde
```

```python
R = rul_roja(1200)
N = rul_negra(1200)
V = rul_verde(1200)
```

```python
# Graficar las probabilidades
plt.plot(R, label="Rojo", color="red")
plt.plot(N, label="Negro", color="black")
plt.plot(V, label="Verde", color="green")

# Agregar leyenda y mostrar la gráfica
plt.legend()
plt.title("Probabilidades de los colores en la ruleta")
plt.xlabel("Número de simulaciones")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```

```python
#Calculamos media, varianza y desviación estándar de cada función

mediaR=np.mean(R)
varR=np.var(R)
devR=np.std(R)

mediaN=np.mean(N)
varN=np.var(N)
devN=np.std(N)

mediaV=np.mean(V)
varV=np.var(V)
devV=np.std(V)
```

```python
print("La media es:" ,mediaR, varR, devR) #Imprimimos los valores
print(mediaN, varN, devN)
print(mediaV, varV, devV)
```

```python
#Establecemos el +- del error
a=mediaR+devR
b=mediaR-devR

c=mediaN+devN
d=mediaN-devN

e=mediaV+devV
f=mediaV-devV
```

```python
plt.axhline(a, color="black", linestyle="--")
plt.axhline(b, color="black", linestyle="--")
plt.axhline(mediaR, color="c", linestyle="--")
plt.plot(R, "r-", label=mediaR)
plt.legend()
plt.show()
```

```python
plt.axhline(c, color="red", linestyle="--")
plt.axhline(d, color="red", linestyle="--")
plt.axhline(mediaN, color="c", linestyle="--")
plt.plot(N, "k-", label=mediaN)
plt.legend()
plt.show()
```

### ¿Qué distribución es?

Aparentemente es una distribución normal. Procederemos a verificar si es una normal:
Usaremos la prueba $K^2$ de D'Agostino:

La prueba $K^2$ de D'Agostino calcula curtosis y asimetría a partir de los datos, para determinar si la distribución de datos se aparta de la distribución normal.
Una característica de la función de distribución normal es que la media, moda y mediana son iguales.

```python
from numpy.random import seed, randn
import matplotlib.pyplot as plt
from scipy.stats import normaltest
# Configuro la semilla aleatoria
seed(1993)
# Genero 100 muestras
data = RR
# Test de DAgostino
stat, p = normaltest(data)
print('Estadisticos=%.3f, p=%.3f' % (stat, p))
# Interpretación
alpha = 0.05
if p > alpha:
   print('La muestra parece Gaussiana o Normal (no se rechaza la hipótesis nula H0)')
else:
   print('La muestra no parece Gaussiana o Normal(se rechaza la hipótesis nula H0)')
```

Respondiendo a la pregunta:

¿Cuál cree usted que es la mejor estrategia?

Considerando las gráficas, las medias y la desviación estándar. Podemos inferir que la mejor estrategia 1 es mejor.

```python
import random
import matplotlib.pyplot as plt

# Definir las probabilidades acumuladas para cada color en la ruleta
probabilidades_acumuladas = {
    'rojo': 18 / 38,
    'negro': 36 / 38,
    'verde': 1 / 38  # 0 y 00 son verdes
}

# Función para simular un giro de la ruleta usando transformada inversa
def simular_giro_ruleta():
    # Generar número aleatorio uniforme entre 0 y 1
    u = random.random()

    # Determinar el resultado de la ruleta según las probabilidades acumuladas
    if u < probabilidades_acumuladas['rojo']:
        return 'rojo'
    elif u < probabilidades_acumuladas['negro']:
        return 'negro'
    else:
        return 'verde'

# Simular juego de la ruleta
def simular_juego_ruleta(num_simulaciones):
    saldo_inicial = 200
    saldo_actual_simple = saldo_inicial
    saldo_actual_martingala = saldo_inicial

    resultados_simple = []
    resultados_martingala = []

    for _ in range(num_simulaciones):
        # Jugar con estrategia simple (apuesta fija a rojo)
        resultado_simple = simular_giro_ruleta()
        if resultado_simple == 'rojo':
            saldo_actual_simple += 1
            resultados_simple.append('ganar')
        else:
            saldo_actual_simple -= 1
            resultados_simple.append('perder')

        # Jugar con estrategia de Martingala
        apuesta = 1
        while True:
            resultado_martingala = simular_giro_ruleta()
            if resultado_martingala == 'rojo':
                saldo_actual_martingala += apuesta
                resultados_martingala.append('ganar')
                break
            else:
                saldo_actual_martingala -= apuesta
                resultados_martingala.append('perder')
                apuesta *= 2
                if apuesta > 500:
                    apuesta = 1  # Resetear apuesta si llega al límite máximo de 500

        # Resetear saldos si llegan a cero
        if saldo_actual_simple <= 0:
            saldo_actual_simple = saldo_inicial
        if saldo_actual_martingala <= 0:
            saldo_actual_martingala = saldo_inicial

    return resultados_simple, resultados_martingala

# Número de simulaciones
num_simulaciones = 1000

# Simular el juego para ambas estrategias
resultados_simple, resultados_martingala = simular_juego_ruleta(num_simulaciones)

# Contar las veces que se gana y se pierde
ganadas_simple = resultados_simple.count('ganar')
perdidas_simple = resultados_simple.count('perder')

ganadas_martingala = resultados_martingala.count('ganar')
perdidas_martingala = resultados_martingala.count('perder')

# Graficar los resultados
labels = ['Ganadas', 'Perdidas']
simple = [ganadas_simple, perdidas_simple]
martingala = [ganadas_martingala, perdidas_martingala]

x = range(len(labels))

plt.figure(figsize=(10, 6))

plt.bar(x, simple, width=0.4, label='Estrategia 1', align='center', color='blue', alpha=0.7)
plt.bar(x, martingala, width=0.4, label='Estrategia 2', align='edge', color='red', alpha=0.7)

plt.xlabel('Resultado')
plt.ylabel('Cantidad')
plt.title('Resultados de Simulación de Ruleta')
plt.xticks(x, labels)
plt.legend()

plt.tight_layout()
plt.show()
```
