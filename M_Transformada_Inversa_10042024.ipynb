# Generación de Variables aleatorias
## Funciones continuas
## Método de la transformada inversa

#Teorema
Sea $U$ una variable uniforme en $(0,1)$. Para cualquier función de distribución continua F invertible, la variable aleatoria *$X$* definida como:

$$X = F^{1}(U) (\text{función inversa } F^{-1})$$

Tiene distribución $F$

/// Recordar: $F^{-1}$ se define como el valor de $x$ tal que $F(x) = u$ ///

# *Demostración*
Sea Fx la función de distribución de X = F¹(U). Entonces

$$Fx(x) = \\ P(X≤x) = P(F¹(U)\leq x) \\ \rightarrow P \left[F(F^{1}(U))≤F(x) \right] = P(U≤F(x)) \\=F(x)$$

Graficamos la función acumulativa de la distribución exponencial

## Ejemplo
Calcular la transformada inversa:
$$F(x)$ = 1 - e^{-λx}$
Para resolver $X$, resolvemos para $x$ en la ecuación:

$$ \text{Sea } F(x)=R \\ \Rightarrow 1 - e^{-λx} = R \\  1 - R = e^{-λx}$$
Aplicamos logaritmo natural para despejar $x$:
$$In(1-R) = - \lambda x \\ \Rightarrow x= - \frac{In(1-R)}{\lambda}$$
Hacemos un cambio de variable, donde $R=u$
$$x= - \frac{In(1-u)}{\lambda}$$

```python
# Importamos las librerías que usaremos
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def expo(k,n): #Definimos una función para generar la distribución exponencial
 L=-1/k #Creamos una variable para el cociente de x, donde k=lambda
 l=[] #Definimos una lista vacía donde guardaremos el valor de x
 for i in range(n):
  u=rnd.random() #Creará números aleatorios entre 0 y 1
  x=L*np.log(u) #Aplicamos la fórmula para obtener x
  l.append(x) #Guardamos el valor de x en la lista
 return l #Pedimos que nos regrese la lista
```

```python
expo(10,5) #Probamos si la función funciona
```

```python
G=expo (0.5,1000000) #Generamos la distribución exponencial
mu=np.mean ( G ) #Calcular la media
sigma2=np.var ( G ) #Calcular la varianza
```

```python
# Imprimimos el valor de la media y de la varianza
print ( mu )
print ( sigma2 )
```

```python
plt.hist ( G ,  bins= 300 ,  color= "orange" ) #Gráfica de la función de densidad
plt.title ( "Función de Densidad" )
```

```python
plt.plot(G) #Está graficando la secuencia de los números aleatorios)
plt.title( "Secuencia")
```

```python
#Esta función nos permite calcular la función acumulativa
def cdf(muestra):
  n=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y=cdf(G)
```

```python
plt.step(X,Y) #El Rango de la función es de 0 a 1
#El dominio es de 0 a infinito
#Es creciente y monótona
plt.title("Función acumulativa")
plt.axhline(1.0000001, color="red",linestyle = "dashed")
#La función cumple que si a<b → F(a)<F(b)
```
